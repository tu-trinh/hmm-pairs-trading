```python
import numpy as np
import pandas as pd
import statsmodels
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import coint
```

```python
# Stock_values will be the predicted stock values in a dataframe, dates as indices and ticks as columns
def find_cointegrated_pairs(stock_values, threshold):
    n = stock_values.shape[1]
    score_matrix = np.zeros((n, n))
    pvalue_matrix = np.ones((n, n))
    keys = stock_values.keys()
    pairs = []
    for i in range(n):
        for j in range(i+1, n):
            S1 = stock_values[keys[i]]
            S2 = stock_values[keys[j]]
            result = coint(S1, S2)
            score = result[0]
            pvalue = result[1]
            score_matrix[i, j] = score
            pvalue_matrix[i, j] = pvalue
            if pvalue < threshold:
                pairs.append((keys[i], keys[j]))
    return score_matrix, pvalue_matrix, pairs
```

```python
# Heatmap
def heatmap(stock_values, threshold):
    scores, pvalues, pairs = find_cointegrated_pairs(stock_values, threshold)
    sns.heatmap(pvalues, xticklabels = stock_values.columns.values, yticklabels = stock_values.columns.values,
                cmap='mako', mask = (pvalues >= 0.98))
    plt.show()
    print(pairs)
```

```python
def zscore(arr):
    return (arr - arr.mean()) / np.std(arr)
```

```python
class PairsTrade():
    def __init__(self, z_window, stats_window, threshold):
        self.z_window = z_window
        self.stats_window = stats_window
        self.threshold = threshold
        self.z_scores = None
        self.ratios = None
        
    def train(self, stock1, stock2):
        if self.z_window == 0 or self.stats_window == 0:
            return 0
        ratios = stock1 / stock2
        mavg1 = ratios.rolling(window = self.z_window, center = False).mean()
#         print(mavg1)
        mavg2 = ratios.rolling(window = self.stats_window, center = False).mean()
        std = ratios.rolling(window = self.stats_window, center = False).std()
        z_score = (mavg1 - mavg2) / std
#         print(z_score)
        self.z_scores = z_score
    
    def trade(self, stock1, stock2):
        money = 0
        count1 = 0
        count2 = 0
        self.ratios = stock1 / stock2
        for i in range(len(self.ratios)):
            # Sell short if the z-score is > thresh
            if self.z_scores[i] > self.threshold:
                money += stock1[i] - stock2[i] * self.ratios[i]
                count1 -= 1
                count2 += ratios[i]
            # Buy long if the z-score is < thresh
            elif self.z_scores[i] < -self.threshold:
                money -= stock1[i] - stock2[i] * self.ratios[i]
                count1 += 1
                count2 -= self.ratios[i]
            # Clear positions if the z-score is in the between thresh
            elif abs(self.z_scores[i]) < self.threshold / 2:
                money += count1 * stock1[i] + stock2[i] * count2
                count1 = 0
                count2 = 0
        return money
```

```python
# Train solely on test companies for now
z_windows = [5]
stats_windows = [30, 40, 50]
thresholds = [0.5, 1, 1.25]
profits = []
for z_window in z_windows:
    for stats_window in stats_windows:
        for threshold in thresholds:
            trader = PairsTrade(z_window, stats_window, threshold)
            trader.train(tech["FB"], tech["ZM"])
            profits.append((z_window, stats_window, threshold,
                            trader.trade(tech["FB"], tech["ZM"])))
#             print("Finished testing z_window of {}, stats_window of {}, and threshold of {}"
#                   .format(z_window, stats_window, threshold))
best_combo = max(profits, key = lambda p : p[3])
print("Best combo: z_window = {}, stats_window = {}, z-score threshold = {}"
      .format(best_combo[0], best_combo[1], best_combo[2]))
```

```python
# Plot the ratios and buy and sell signals from z score
trader = PairsTrade(5, 30, 0.5)
split_idx = 25
tech.drop(columns = ["AAPL"], inplace = True)
trader.train(tech["FB"], tech["ZM"])
profit = trader.trade(tech["FB"], tech["ZM"])
print("Profit:", profit)

plt.figure(figsize=(15,7))
relevant = tech
relevant.plot()
buy = relevant.copy()
sell = relevant.copy()
buy[trader.z_scores > trader.threshold] = 0
sell[trader.z_scores < trader.threshold] = 0
buy.plot(color='g', linestyle='None', marker='^')
sell.plot(color='r', linestyle='None', marker='^')
x1,x2,y1,y2 = plt.axis()
plt.axis((x1,x2,trader.ratios.min(),trader.ratios.max()))
plt.legend(['Ratio', 'Buy Signal', 'Sell Signal'])
plt.show()
```
